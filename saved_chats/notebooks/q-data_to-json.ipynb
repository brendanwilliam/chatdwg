```python
import pandas as pd
import json

# read in the data
df = pd.read_csv('spotthebot-questions.csv')

# create an empty list to store the results
results = []

# iterate over each row in the dataframe
for index, row in df.iterrows():

    # create a dictionary for each answer
    answer1 = {
        'answerContent': row['answer_1'],
        'chatGPTAnswer': row['isGPT_1']
    }
    answer2 = {
        'answerContent': row['answer_2'],
        'chatGPTAnswer': row['isGPT_2']
    }
    answer3 = {
        'answerContent': row['answer_3'],
        'chatGPTAnswer': row['isGPT_3']
    }

    # create a dictionary for the question and answers
    question_dict = {
        'question': row['question'],
        'answers': [answer1, answer2, answer3]
    }

    # append the question dictionary to the results list
    results.append(question_dict)

# print the JSON string
with open('spotthebot-questions.json', 'w') as f:
    json.dump(results, f, indent=4)
```
